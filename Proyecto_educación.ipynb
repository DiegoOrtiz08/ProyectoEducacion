### **1.3.2. Indicadores (reprobación, deserción, cobertura_neta), promedio por año**

Durante la exploración de los datos, observamos que, a pesar de que la _cobertura neta_ era superior al $80\%$ y de que los valores de _deserción_ disminuian (lo que es consistente con el sentido común), los valores de _reprobación_ se van elevando. ### **1.4.1. Prueba de normalidad**
---

Para esta prueba queremos saber si los datos obtenidos por año provienen de una distribución normal. ¿Para qué hacemos esto? Debido a que una gran parte de estadístico asumen normalidad, debemos realizar las pruebas correspondientes para verificar tal característica.

```python
# Seleccionamos los datos por el año 2019, 2020 y 2021
_2019 =  df_dep[df_dep['AÑO'] == 2019]
_2020 =  df_dep[df_dep['AÑO'] == 2020]
_2021 =  df_dep[df_dep['AÑO'] == 2021]
```

```python
# Definimos una función para almacenar los estadísticos y p-valor de las características por año
def normalidad(data):
  '''
  Esta función construye un DataFrame de pandas donde almacenará estadístico y p-valor
  de la prueba stats.shapiro

  :input: un conjunto de datos, data
  :output: tabla con el siguiente encabezado    |index|'Variable'|'Estadístico'|'p-valor|
  '''
  tabla_var_normal = pd.DataFrame({'Variable': [],'Estadístico':[], 'p-valor': []})

  for columna in (data.loc[:,'POBLACIÓN_5_16':]).columns:
      est_temp, pval_temp = stats.shapiro(data[columna])
      tabla_var_normal = pd.concat([tabla_var_normal,
                                    pd.Series({'Variable': columna, 'Estadístico': est_temp, 'p-valor': pval_temp}).to_frame().T],
                                    ignore_index=True)
  return tabla_var_normal
```

```python
#tabla de normalidad para los años 2019,2020 y 2021
tabla_norm_2019 = normalidad(_2019)
tabla_norm_2020 = normalidad(_2020)
tabla_norm_2021 = normalidad(_2021)
```

```python
#Variables cuyo p-valor es mayor al 0.05 y por tanto no satisfacen las condiciones
#de normalidad, de acuerdo con la prueba stats.shapiro(data)

#2019
print(tabla_norm_2019[tabla_norm_2019['p-valor']>0.05],'\n')
#2020
print(tabla_norm_2020[tabla_norm_2020['p-valor']>0.05],'\n')
#2021
print(tabla_norm_2021[tabla_norm_2021['p-valor']>0.05])
```

###**1.4.2. Diferencia en las medias poblacionales**

La siguiente tabla muestra si los estudiantes de un año tuvieron una aprobación diferente a los estudiantes del año consecutivo. De este modo,
+ $H_0$ sostiene que no hubo diferencia en la aprobación entre dos años consecutivos, mientras que nuestra hipótesis alternativa
+ $H_1$ sostiene que sí hubo diferencia en cuanto a la aprobación.

```python
data = []
for año in range(2011, 2022):
    if año + 1 <= 2021:
        z_score, p_valor = sm.stats.ztest(df_dep[df_dep['AÑO'] == año]['APROBACIÓN'], df_dep[df_dep['AÑO'] == año + 1]['APROBACIÓN'],  # Se toman las dos muestras a partir del año y la aprobación media
                                          value=0,  # La dif=0 es la hipótesis nula
                                          alternative='two-sided'  # dos colas
                                          )
        if p_valor < 0.05:
            p_v = "Rechazamos la hipótesis nula"
        else:
            p_v = "No rechazamos la hipótesis nula"
        data.append({'Años': f"{año} - {año + 1}", 'Estadístico': z_score, 'p-valor': p_valor, 'Decisión': p_v})

df = pd.DataFrame(data)
df
```

En los casos en los que el p-valor es menor al nivel de significancia $(5\%)$, podremos rechazar la hipótesis nula. Esto significa que hay evidencia que sugiere que los estudiantes del _2013_ a _2014_, del _2014_ a _2015_ y del _2020_ a _2021_ tienen un aprobación diferente entre sí, mientras que entre los otros años consecutivos que aparecen en la tabla no hay evidencia para sostener que hubo diferencia en la aprobación.
